```python
# Function to find team name against who the batsmans score the max runs
def run_scored(batsman_name):
  bm = delivery[delivery['batsman']==batsman_name]
  return bm.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
run_scored('MS Dhoni')
```

```python
#isin() function
# Find the most destructive death over batsman in the history of IPL
# We do that on the basis of strike rate (no. of runs/no. of balls)*100
# Condition - Batsman atleast played 200 balls in over 16-20
```

```python
death_over = delivery[delivery['over']>15]
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
batsman_list = all_batsman[x].index.tolist()
```

```python
final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr
```
