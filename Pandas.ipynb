```python
delivery = pd.read_csv('deliveries.csv')
delivery.head()
```

```python
runs = delivery.groupby('batsman')
```

```python
runs['batsman_runs'].sum().sort_values(ascending=False)
```

```python
mask = delivery['batsman_runs']==4
new_delivery = delivery[mask]
```

```python
new_delivery.shape[0]
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
```

```python
vk = delivery[delivery['batsman']=='V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
# Function to find team name against who the batsmans score the max runs
def run_scored(batsman_name):
  bm = delivery[delivery['batsman']==batsman_name]
  return bm.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
run_scored('MS Dhoni')
```
